```python
import requests 
from bs4 import BeautifulSoup
import pandas as pd
from IPython.core.display import display,HTML
```

# Fetching Legendary items in the game Diablo 3

```python
url='https://www.diablowiki.net/Legendary'
```

```python
soup=BeautifulSoup(requests.get(url).content,"html.parser")
```

```python
print(soup.prettify)
```

```python
tables=soup.find_all('div',attrs={'class':'NavTabContent','title':['Armor1','Armor2','Weapons1','Weapons2']})
```

```python
name = []
urls = []
    
for data in table:
    data.select('td')
    for link in data.select('td'):
        link.find_all('a')
        for text in link.find_all('a'):
            if text.text.startswith("Legendary") or text.text.startswith("Crafted"):
                continue
            url = "https://www.diablowiki.net/"+text.text
            name.append(text.text)
            urls.append(url)

print(len(name))
print(len(urls))
```

```python
print(name[:10])
print('-------------------------------------------------------------')
print(urls[:10])
```

# Fetching description and image link from each legendary item

```python
img_link=[]
items_description=[]

for i in urls:
    soup=BeautifulSoup(requests.get(i).content,"html.parser")
    soup.img['src']
    img_url = "https://www.diablowiki.net/"+soup.img['src']
    img_link.append(img_url)
    items_description.append(soup.p.text.rstrip())
    print(i,'Done!')
```

```python
print(len(img_link))
print(len(items_description))
```

```python
print(img_link[:10])
print('-------------------')
print(items_description[:10])
```

# Creating dataframe

```python
df=pd.DataFrame({'Legendary_item':name,'Description':items_description,'Image_link':img_link,'Link':urls})
```

```python
df.set_index('Legendary_item',inplace=True)
```

# Adding item image into dataframe

```python
def path_to_image_html(path):
    return '<img src="'+ path + '" width="60" >'

pd.set_option('display.max_colwidth', None)

image_cols = ['Image_link']

format_dict = {}
for image_col in image_cols:
    format_dict[image_col] = path_to_image_html


display(HTML(df.to_html(escape=False ,formatters=format_dict)))
```
